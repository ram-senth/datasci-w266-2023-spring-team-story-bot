```python
book_list = booklist_string.split(",")
i = 0
for item in book_list:
    book_list[i] = item.lstrip()
    i += 1
book_list
```

```python
book_dict = dict()
with open('cleaned_merged_fairy_tales_without_eos.txt','r') as f:
    lines = f.readlines()
    for book in book_list:
        for row in lines:
            if book in book_dict:
                continue
            #if row.find(book) != -1:
            if re.search(book, row, re.IGNORECASE):
                book_dict[book] = lines.index(row)
book_dict
```
